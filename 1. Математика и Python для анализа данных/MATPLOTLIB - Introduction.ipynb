# Совмеетное использование NumPy, SciPy, Matplotlib
---

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

Модуль SciPy interpalte позволяет выполнять интерполяцию

$$ y = e^{-x / 3} $$

```python
x = np.arange(0, 10, 2)
y = np.exp(-x/3.0) + np.random.randn(len(x)) * 0.05 
# Необходимо указывать значение 3.0 вместо 3, чтобы не происходило деление нацело
# Данная часть функции (np.random.randn(len(x)) * 0.05) моделируте шум в данных

print(x[:5])
print(y[:5])
```

### Метод .interp1d() позволяет выполнять интерполяцию

По умолчанию параметр kind метода .interp1d() равен 'linear'.  Можно указывать другие значения данного параметра, например, 'quadratic'.

```python
f = interpolate.interp1d(x, y, kind='quadratic')
xnew = np.arange(0, 8, 0.1)
ynew = f(xnew)
```

```python
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.show()
```

Синие точки - известные значения.

Оранжевая линия - приближение.
